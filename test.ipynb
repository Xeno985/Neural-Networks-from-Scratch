```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from layers import dense,ReLU
from activation import Tanh,sigmoid
from loss import Quadloss
from model import sequential

X = np.reshape([[0, 0], [0, 1], [1, 0], [1, 1],[0,2],[2,1]], (6, 2, 1))
Y = np.reshape([[0], [1], [1], [0],[1.25],[1.5]], (6, 1, 1))

network =sequential( [
    dense(2, 3),
    sigmoid(),
    dense(3, 5),
    Tanh(),
    dense(5,8),
    sigmoid(),
    dense(8,6),
    Tanh(),
    dense(6,7),
    sigmoid(),
    dense(7,1),
    sigmoid()
], Quadloss)


network.train( X, Y, epochs=1200 )


```
